```python
#Sum of n numbers
sum = 0
for i in range(0,10):
    num = int(input('Enter number : '))
    sum += num
print('Total Sum : ',sum)
```

```python
# Fibonacci Series upto a limit
limit = int(input('Enter how many numbers you want : '))
f,l = 0,1
s = 0
s = f+l
print('First ',limit,' numbers in Fibonacci Series are : ')
print(f, l, end = " ")
for i in range(limit-2):
    print(s, end=" ")
    f = l 
    l = s
    s = f+l
```

```python
# Prime numbers upto a limit

lwr_limit = 10
upr_limit = 100
print("Prime numbers between ",lwr_limit," and ",upr_limit," are : ")
for num in range(lwr_limit, upr_limit + 1):
    if num > 1:
        for i in range(2,num):
            if (num % i) == 0:
                break
            else:
                print(num)
```

```python
# Check whether a number is Armstrong number

num = int(input('Enter a number : '))
temp = num
s = 0
while num > 0:
    r = num % 10
    s = s + r*r*r
    num = num // 10
if s == temp:
    print(temp, ' is Armstrong number')
else:
    print(temp, ' is not Armstrong number')
```
